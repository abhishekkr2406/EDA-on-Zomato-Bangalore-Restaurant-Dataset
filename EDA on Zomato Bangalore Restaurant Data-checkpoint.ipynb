## Zomato Bangalore Restaurants Dataset Download

```python
data_filename = './zomato-bangalore-restaurants/zomato.csv'
```

## Data preparation & Cleaning
1. Loading the data.
2. Brief look over the information given.
3. Drop unnecessary columns.
4. Look for missing and duplicate values.
5. Replace the missing values with data having most count.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv(data_filename)
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.columns
```

```python
# Dropping unnecessary columns.

df.drop(['url', 'address', 'reviews_list'], axis=1, inplace = True)
```

```python
df.isna().sum()
```

```python
df['rate'].unique()
```

```python
# Some of the unique values in rate are 'NEW' or '-'. We will replace these values with '0' and then drop them.

df['rate'] = df['rate'].apply(lambda x: x if x not in ['NEW', '-'] else '0')
df['rate'].unique()
```

```python
# Removing all the data having null value in rate.

df.dropna(subset=['rate'], how='any', inplace=True)
df.reset_index(drop=True)
```

```python
# Change the datatype of rate from str to float.

df['rate'] = [float(str(value).split('/')[0]) for value in df['rate']]
```

```python
# Change the data type of approx_cost.

df['approx_cost(for two people)'] = [float(str(value).replace(',','')) for value in df['approx_cost(for two people)']]
df['approx_cost(for two people)'].unique()
```

```python
df.isna().sum()
```

```python
# Rename the column 'approx_cost(for two people)' to 'average_cost' and replace all tha null values with the mean average_cost.

df.rename(columns = {'approx_cost(for two people)':'average_cost'}, inplace = True)
df['average_cost'].fillna(df['average_cost'].mean(), inplace=True)
df['average_cost'] = df['average_cost'].astype(int)
```

```python
df.describe()
```

```python
# Average rating of each restaurant

df_rating = df.groupby('name')['rate'].mean().to_frame()
df_rating = df_rating.reset_index()
df_rating.columns = ['restaurant', 'rating']
df_rating.head(15)
```

```python
sns.distplot(df_rating['rating'])
```

- From the above distribution, we can see that most of the restaurants have been given ratings between 3.0 & 4.0.
- There are very few restaurants whose rating is above 4.0 and even fewer above 4.5.

```python
# Restaurants under the budget of 500.

df_under500 = df[df.average_cost < 500]
df_under500.reset_index(inplace=True)
df_under500['average_cost'].value_counts()
```

```python
sns.barplot(x=df_under500['average_cost'].value_counts().index, y=df_under500['average_cost'].value_counts())
plt.xticks(rotation='vertical')
plt.xlabel('Average Cost')
plt.ylabel('No. of Restaurants')
plt.show()
```

```python
# Restaurants under budget of Rs.500 with Rating>4.0

df_rate500 = df_under500[df_under500.rate > 4]

fig, ax = plt.subplots(figsize=(20, 5))

sns.barplot(x=df_rate500['location'].value_counts().index, y=df_rate500['location'].value_counts())
plt.xticks(rotation='vertical')
plt.xlabel('Locations')
plt.ylabel('No. of Restaurants')
plt.show()
```
